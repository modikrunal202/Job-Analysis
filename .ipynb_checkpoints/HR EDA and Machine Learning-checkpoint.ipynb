## All models in a single function

```python
def Classification_model(model, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=111)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    accuracy = round(accuracy_score(y_test, pred)*100, 2)
    return accuracy
```

```python
models=["Random Forest","Decision Tree","Gaussian Naive Bays","K-Nearest Neighbors","Logistic Regression","Linear SVC"]
Classification_model_list = [RandomForestClassifier(n_estimators=200),DecisionTreeClassifier(),
                             GaussianNB(),KNeighborsClassifier(n_neighbors=1)
                             ,LogisticRegression(),SVC()]
Model_accuracy = []
X = hr_data.drop('left', axis=1)
y = hr_data['left']

for model in Classification_model_list:
    Accuracy = Classification_model(model, X, y)
    Model_accuracy.append(Accuracy)
```

```python
Accuracy_with_all_features_function = pd.DataFrame(
    {
        "Classification Model" :models,
         "Accuracy with all features": Model_accuracy
     
    })
Accuracy_with_all_features_function = Accuracy_with_all_features_function.sort_values(by="Accuracy with all features"
                                                                                      ,ascending=False).reset_index(drop=True)
```

```python
Accuracy_with_all_features_function
```

```python
# Lets try with Important features
Model_Acc = []
importan_fet = Important_features = ['satisfaction_level','number_project','time_spend_company','average_montly_hours',
                                     'last_evaluation']
imp_fet = hr_data[importan_fet]
pred_var = hr_data['left']
for model in Classification_model_list:
    Accuracy=Classification_model(model,imp_fet,pred_var) # Just instead of all features give only important features
    Model_Acc.append(Accuracy)
```

```python
Accuracy_with_important_feature = pd.DataFrame(
    {
        "Classification Model" :models,
         "Accuracy with Important features": Model_Acc
     
    })
Accuracy_with_important_feature = Accuracy_with_important_feature.sort_values(by="Accuracy with Important features"
                                                                                      ,ascending=False).reset_index(drop=True)
```

## 3.2 Machine Learning Models With Cross Validation
1. In this we will do the cross validation with the models to get there mean accuracy
2. From 3.1 we came to know that by using important features there is increase in the accuracy for mostly all models so in this we will go only with important features

```python
# This is used for to caculate the score of cross validation by using Kfold
from sklearn.model_selection import cross_val_score

def Classification_model_CV(model, X, y):
    '''This function will return accuracy for different suplied models.
    ** Parameters **
    1) model : In this parameter list of different classification models are given.
    2) X : In this parameter dataframe with all feature is supplied
    3) y: In this parameter dataframe with predict variable is provided
    '''
    x_data = X
    y_data = y
    scores = cross_val_score(model, x_data, y_data, scoring = 'accuracy', cv=10)
    accuracy = scores.mean()
    return accuracy
```

```python
Model_accuracy = []
for model in Classification_model_list:
    Accuracy = Classification_model_CV(model,imp_fet, pred_var)
    Model_accuracy.append(round(Accuracy*100, 2))
```

```python
Accuracy_with_cross_val = pd.DataFrame(
    {
        "Classification Model" :models,
         "Accuracy with Cross Validation": Model_accuracy
    })
Accuracy_with_cross_val = Accuracy_with_cross_val.sort_values(by="Accuracy with Cross Validation"
                                                            ,ascending=False).reset_index(drop=True)
Accuracy_with_cross_val
```

```python
comaprision = pd.merge(pd.merge(Accuracy_with_all_features_function, Accuracy_with_important_feature, on='Classification Model')
                      , Accuracy_with_cross_val, on = 'Classification Model')
```

```python
comaprision = comaprision.loc[:,['Classification Model', 'Accuracy with all features', 'Accuracy with Important features', 
                                 'Accuracy with Cross Validation']]
comaprision
```

# 3.3 Machine Learning Models With Parameter tuning
1. In this we will use Grid SearchCV to find the best parameter for a model
2. in this we will use important features.

```python
#Import 
from sklearn.model_selection import GridSearchCV

def Classification_model_GridSearchCV(model, X, y, params):
    x_data = X
    y_data = y
    clf = GridSearchCV(model,params,scoring="accuracy",cv=5)
    clf.fit(x_data, y_data)
    best_score = round(clf.best_score_*100, 2)
    print('Best Score is : ', best_score)
    print('\n')
    print('Best estimator  is :', clf.best_estimator_)
    
    return best_score
```
